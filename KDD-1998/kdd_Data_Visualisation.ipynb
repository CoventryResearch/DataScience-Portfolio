```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
cd Dropbox/Portfolio/DataScience-Portfolio/KDD-1998
```

```python
# load data
kdd = pd.read_csv('data_reg2.csv')
```

```python
kdd.TARGET_D.describe()
```

```python
fig = plt.figure()
plt.hist(kdd.TARGET_D, bins = 50)
plt.title('Donations distribution')
plt.xlabel('Donation in USD')
plt.ylabel('Number of People')
```

```python
fig = plt.figure()
plt.hist(np.log(kdd.TARGET_D + 1), bins = 50)
plt.title('Donations distribution')
plt.xlabel('Donation in USD, log')
plt.ylabel('Number of People')
```

```python
imp_feat = ['LASTGIFT', 'RFA_2A_G', 'AVGGIFT','RFA_2_L1G', 'MAXRAMNT', 'RFA_2F', 'RFA_3_A1G', 'RFA_4_A1G','RFA_2A_D',
            'MDMAUD_A_C']
```

```python
for i in range(0,10,2):
    
    f, (ax1,ax2) = plt.subplots(1,2, figsize = [12,4])
    ax1.plot(kdd[imp_feat[i]], kdd.TARGET_D, 'o')
    ax1.set_xlabel(imp_feat[i])
    ax1.set_ylabel('Donation Amount (USD)')
    
    ax2.plot(kdd[imp_feat[i+1]], kdd.TARGET_D, 'o')
    ax2.set_xlabel(imp_feat[i+1])
    ax2.set_ylabel('Affair Rate')
```

```python
for i in range(0,10,2):
    
    f, (ax1,ax2) = plt.subplots(1,2, figsize = [12,4])
    ax1.plot(np.log(kdd[imp_feat[i]]+1), np.log(kdd.TARGET_D), 'o')
    ax1.set_xlabel(imp_feat[i])
    ax1.set_ylabel('Donation Amount (USD)')
    
    ax2.plot(np.log(kdd[imp_feat[i+1]]+1), np.log(kdd.TARGET_D), 'o')
    ax2.set_xlabel(imp_feat[i+1])
    ax2.set_ylabel('Affair Rate')
```

```python
plt.plot(kdd.Age, kdd.TARGET_D, 'o')
plt.title('Age on Donations')
plt.xlabel('Age')
plt.ylabel('Donations')
```

```python
plt.plot(np.log(kdd.Age), np.log(kdd.TARGET_D), 'o')
plt.title('Age on Donations')
plt.xlabel('Age')
plt.ylabel('Donations')
```
